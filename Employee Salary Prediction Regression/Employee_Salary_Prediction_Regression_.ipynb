```python
# Feature Engineering
```

```python
# Label Encoding:
```

```python
# Label Encoding:
from sklearn.preprocessing import LabelEncoder
Label_Encoder = LabelEncoder()

# Use 'Gender' instead of 'Ge' to access the correct column
sal_data['Gender_Encode']= Label_Encoder.fit_transform(sal_data['Gender'])

sal_data['Degree_Encode']= Label_Encoder.fit_transform(sal_data['Degree'])

sal_data['Job_Title_Encode']= Label_Encoder.fit_transform(sal_data['Job_Title'])
```

```python
# Data after label Encoding:
```

```python
sal_data.head()
```

```python
# Feature Scalling
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler= StandardScaler()
```

```python
sal_data['Age_scaled']= std_scaler.fit_transform(sal_data[['Age']])
# sal_data['Experience_years_scaled']= std_scaler.fit_transform(sal_data[['Exprience_years']]) # Incorrect spelling
sal_data['Experience_years_scaled']= std_scaler.fit_transform(sal_data[['Experience_years']]) # Correct spelling
```

```python
sal_data.head()
```

```python
# Dependent and independent
```

```python
# Dependent and independent

# Define x by dropping the original columns used for scaling/encoding and the target variable
x = sal_data.drop(['Age', 'Gender', 'Degree', 'Job_Title', 'Experience_years', 'Salary'], axis=1)

# Define y as the target variable
y = sal_data['Salary']
```

```python
x.head()
```

```python
# Splitting the data into traning and Testing:
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_tarin,  y_test= train_test_split(x, y, test_size= 0.2, random_state=42)
x_train.head()
```

```python
x_train.shape, y_tarin.shape
```

```python
x_test.shape, y_test.shape
```

```python
# Modal Development
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Linear_regression_modal= LinearRegression()
```

```python
# Modal Traning:
```

```python
Linear_regression_modal.fit(x_train, y_tarin)
```

```python
# Modal Predection:
```

```python
y_pred_lr= Linear_regression_modal.predict(x_test)
y_pred_lr
```

```python
df= pd.DataFrame({'y_Actual' : y_test, 'y_predicted': y_pred_lr})
df['Error']= df['y_Actual']- df['y_predicted']
df['abs_error']= abs(df['Error'])
```

```python

mean_absolute_Error= df['abs_error'].mean()
mean_absolute_Error
```

```python
# Modal Evaluation
```

```python
from sklearn.metrics import accuracy_score, r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Modal Accuracy:
```

```python
r2_score(y_test, y_pred_lr)
```

```python
print(f' Account of the modal = {round(r2_score(y_test, y_pred_lr),4)*100} %')
```

```python
# Mean Absolute Error:
```

```python
round(mean_absolute_error(y_test, y_pred_lr),2)
```

```python
print(f" Mean Absolute Error = {round(mean_absolute_error(y_test, y_pred_lr),2)}")
```

```python
# Mean Squared Error:
```

```python
mse= round(mean_squared_error(y_test, y_pred_lr),2)
mse
```

```python
print(f"Mean Sqaured Error = {round(mean_squared_error(y_test, y_pred_lr),2)}")
```

```python
# Root mean aquared error:
print('Root Mean squared Error (RMSE)=', mse**(0.5))
```
